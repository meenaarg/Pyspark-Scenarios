```python

print("-----GIVEN DATAFRAME-----\n\n\n")

df = spark.createDataFrame([("Renuka1992@gmail.com", "9856765434"), ("anbu.arasu@gmail.com", "9844567788")], ["email", "mobile"])
df.show()

print("-----REQUIRED DATAFRAME-----\n\n\n")


data1 = [
    ("R*******92@gmail.com", "9*******34"),
    ("a*******su@gmail.com", "9*******88")
]

columns1 = ["email", "mobile"]

df1 = spark.createDataFrame(data1, columns1)

df1.show()

print("-----DSL SOLUTION-----\n\n\n")

import pyspark.sql.functions as sf
delimitterdf = df.withColumn("emailbody", sf.split("email", "@")[0])\
    .withColumn("emailend", sf.split("email", "@")[1])\
        .drop("email")

delimitterdf.show()



maskdf = delimitterdf.withColumn("efirst", sf.substring(sf.col('emailbody'),1,1))\
    .withColumn("elast",sf.substring(sf.col('emailbody'),-2,2))\
        .withColumn("emailstart",sf.concat(
            sf.col("efirst"),
            sf.repeat(sf.lit("*"), sf.length(sf.col("emailbody")) - 3),
            sf.col("elast")))\
                .withColumn("mobilefirst",sf.substring(sf.col('mobile'),1,1))\
.withColumn("mobilelast",sf.substring(sf.col('mobile'),-2,2))\
.withColumn("mobile",sf.concat(
sf.col("mobilefirst"),
sf.repeat(sf.lit("*"), 7),
sf.col("mobilelast")))

maskdf.show()

finaldf = maskdf.withColumn("email", sf.concat(sf.col("emailstart"), sf.lit("@"), sf.col("emailend")))\
    .drop("emailbody", "emailend", "efirst", "elast", "emailstart", "mobilefirst", "mobilelast")\
.select("email","mobile")

finaldf.show()
```
