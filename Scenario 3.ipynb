```python

print("-----GIVEN DATAFRAME-----\n\n\n")

food_data = [
    (1, "Veg Biryani"),
    (2, "Veg Fried Rice"),
    (3, "Kaju Fried Rice"),
    (4, "Chicken Biryani"),
    (5, "Chicken Dum Biryani"),
    (6, "Prawns Biryani"),
    (7, "Fish Biryani"),
]
fooddf = spark.createDataFrame(food_data, ["food_id", "food_item"])

rating_data = [
    (1, 5), (2, 3), (3, 4), (4, 4), (5, 5), (6, 4), (7, 4)
]
ratingdf = spark.createDataFrame(rating_data, ["food_id", "rating"])

fooddf.show()
ratingdf.show()

print("\n\n\n-----REQUIRED DATAFRAME-----\n\n\n")

data = [
    (1, "Veg Biryani",         5),
    (2, "Veg Fried Rice",      3),
    (3, "Kaju Fried Rice",     4),
    (4, "Chicken Biryani",     4),
    (5, "Chicken Dum Biryani", 5),
    (6, "Prawns Biryani",      4),
    (7, "Fish Biryani",        4),
]

df1 = spark.createDataFrame(data, ["food_id", "food_item", "rating"])

import pyspark.sql.functions as f


# Add the Stats(out of 5) column using repeat()
finalDF = df1.withColumn("Stats(out of 5)", f.repeat(f.lit("*"), f.col("rating")))

finalDF.show()



print("\n\n\n-----SOLUTION-----\n\n\n")

joindf = fooddf.join(ratingdf,"food_id","left")




starsdf = joindf.withColumn("Stats(out of 5)",f.repeat(f.lit("*"),f.col("rating")))

starsdf.show()


```
