```python
from pyspark.sql.functions import *

print("-----GIVEN DATAFRAME-----\n\n\n")
data = [
    (101, "Eng", 90),
    (101, "Sci", 80),
    (101, "Mat", 95),
    (102, "Eng", 75),
    (102, "Mat", 90),
    (102, "Sci", 85)
]
columns = ["Id", "Subject", "Marks"]

df = spark.createDataFrame(data, columns)
df.show()

print("\n\n\n-----REQUIRED DATAFRAME-----\n\n\n")

req = spark.createDataFrame(
    [(101, 90, 80, 95),
     (102, 75, 85, 90)],
    ["Id", "Eng", "Sci", "Mat"]
)
req.show()


print("\n\n-----SOLUTION-----\n\n")

pivotdf = df.groupBy("Id").pivot("Subject").agg(first("Marks"))
pivotdf.show()

```
