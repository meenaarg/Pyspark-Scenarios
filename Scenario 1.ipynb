```python
from pyspark.sql import *
from pyspark.sql.functions import *

print("-----GIVEN DATAFRAME-----\n\n\n")
data = [
    ('A', 'D', 'D'),
    ('B', 'A', 'A'),
    ('A', 'D', 'A')
]

df = spark.createDataFrame(data).toDF("TeamA", "TeamB", "Won")

df.show()

print("\n\n\n-----REQUIRED DATAFRAME-----\n\n\n")

data1 = [
    ("A", 2),
    ("B", 0),
    ("D", 1)
]

# Define column names
columns = ["TeamName", "won"]

# Create DataFrame
df1 = spark.createDataFrame(data1, columns)

# Display
df1.show()


print("\n\n\n-----SOLUTION-----\n\n\n")

teamadf = df.select(col("TeamA").alias("TeamName"))
teambdf = df.select(col("TeamB").alias("TeamName"))
uniondf = teamadf.union(teambdf).distinct()
uniondf.show()

aggdf = df.groupBy("Won").count()


finaldf = uniondf.join (aggdf, uniondf.TeamName == aggdf.Won, "full").na.fill(0).orderBy("TeamName").drop("won").withColumnRenamed("count", "won")
finaldf.show() 
```
