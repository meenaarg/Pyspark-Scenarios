```python
print("-----GIVEN DATAFRAME-----\n\n\n")

# First DataFrame - df
data = [
    (1, "a", 10000),
    (2, "b", 5000),
    (3, "c", 15000),
    (4, "d", 25000),
    (5, "e", 50000),
    (6, "f", 7000)
]

columns = ["empid", "name", "salary"]

df = spark.createDataFrame(data, columns)
df.show()

print("-----REQUIRED DATAFRAME-----\n\n\n")

# Second DataFrame - df1
data1 = [
    (1, "a", 10000, "Employee"),
    (2, "b", 5000, "Employee"),
    (3, "c", 15000, "Manager"),
    (4, "d", 25000, "Manager"),
    (5, "e", 50000, "Manager"),
    (6, "f", 7000, "Employee")
]

columns1 = ["empid", "name", "salary", "Designation"]

df1 = spark.createDataFrame(data1, columns1)
df1.show()

print("-----DSL SOLUTION-----\n\n\n")
from pyspark.sql.functions import *

desigdf = df.withColumn("Designation", when(col("salary")>10000,"Manager").otherwise("Employee"))
desigdf.show()



print("-----SQL SOLUTION-----\n\n\n")


df.createOrReplaceTempView("df_view")

spark.sql("""
          
          select empid, name, salary,
          case when salary>10000 then "Manager" else "Employee" end as Designation
          from df_view
          
          """).show()

```
