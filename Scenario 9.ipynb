```python
print("-----GIVEN DATAFRAME-----\n\n\n")

data = [
    (1, "Mark Ray", "AB"),
    (2, "Peter Smith", "CD"),
    (1, "Mark Ray", "EF"),
    (2, "Peter Smith", "GH"),
    (2, "Peter Smith", "CD"),
    (3, "Kate", "IJ")
]

columns = ["custid", "custname", "address"]

df = spark.createDataFrame(data, columns)
df.show()



print("-----REQUIRED DATAFRAME-----\n\n\n")

data1 = [
    (1, "Mark Ray", ["EF", "AB"]),
    (2, "Peter Smith", ["CD", "GH"]),
    (3, "Kate", ["IJ"])
]

columns1 = ["custid", "custname", "address"]

df1 = spark.createDataFrame(data1, columns1)
df1.show()


print("-----DSL SOLUTION-----\n\n\n")

from pyspark.sql.functions import collect_set
from pyspark.sql.functions import col


collectdf = df.groupBy("custid","custname")\
        .agg(collect_set("address").alias("alias"))\
            .orderBy("custid")
        

collectdf.show()

##collect_set removes duplicates while collect_list doesnt
##To order elements within array, use the sort_array() with collect()




print("-----SQL SOLUTION-----\n\n\n")


df.createOrReplaceTempView("df_view")

spark.sql(""" 
          
          select custid,custname,
          collect_set(address) as address
          from df_view
          group by custid,custname
          
          """).show()
```
